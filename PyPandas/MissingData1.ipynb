## Handling of Missing Data


```python
import pandas as pd
df = pd.read_csv("weather.csv",date_parser=["day"])

df
```

```python
new_df = df.fillna({
    'Temperature':0,
    'WindSpeed' :0,
    'Event' :'No event'
})
new_df
#Filling specific missing values in cols by specific values
#Replace all NaN values by 0.0 and event by NoEvent
```

Other estimate is to fill the previous data

```python
new_df = df.fillna(method="ffill")
new_df
```

```python
df
```

Fill the missing data by one field ahead(Before Fill)

```python
new1_df=df.fillna(method="bfill")
new1_df
```

```python
new_df=df.fillna(method="bfill", axis="columns")
new_df

# Axis helps in copying the missing data horizontally
```

```python
new1_df=df.fillna(method="ffill")
new1_df
```

```python
new1_df=df.fillna(method="ffill",limit=1)
new1_df
# to fill the NaN values in limited cells
```
