## Handling Missing Data(Replace Fxn)

```python
# Special values in csv file like -9999 etc
import pandas as pd
import numpy as np
df =pd.read_csv('weather_data1.csv')
df
```

```python
new_df = df.replace([-99999,-88888],np.NaN)
new_df
#Replacing -99999 by NaN
```

```python
#Replace values based on cols using Dictionary
new_df =df.replace({
    'temperature': -99999,
    'windspeed': -88888,
    'event': '0'
},np.NaN)
new_df
```

```python
new_df =df.replace({
    -99999: np.NaN,
    -88888: np.NaN,
    'NaN' : 'Sunny'
})
new_df
```

Now remove the units of data by REGEX

Regex are used to replace the values
##### Below we are replacing all alphabets by ' ' using regex but it removes data from event too

```python
new_df =df.replace('[A-Za-z]','',regex=True)
new_df
```

##### Using on specific cols

```python
new_df =df.replace({
    'temperature':'[A-Za-z]',
    'windspeed' : '[A-Za-z]'
},'',regex=True)
new_df
```
