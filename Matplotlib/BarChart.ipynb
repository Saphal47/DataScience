## Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

Simple bar chart showing revenues of major US tech companies

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
plt.bar(company,revenue)
#for bar chart
```

```python
#long way to do
xpos = np.arange(len(company))
xpos
```

```python
plt.ylabel("Revenue(billion)")
plt.title('US Tech Stocks')

plt.bar(xpos,revenue,label="Revenue")
plt.legend()
```

```python
plt.xticks(xpos,company) #replace values in x by string in company
plt.bar(xpos,revenue)
```

Multiple Bars showing revenue and profit of major US tech companies

```python
profit=[40,2,34,12]
```

```python
plt.ylabel("Revenue(billion)")
plt.title('US Tech Stocks')

plt.bar(company,revenue,label="Revenue")
plt.bar(company,profit,label="Profit")
plt.legend()
```

```python
#for SidebySide bars
plt.bar(xpos-0.2,revenue,label="Revenue")
plt.bar(xpos+0.2,profit,label="Profit")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

```python
plt.bar(xpos-0.2,revenue, width=0.4, label="Revenue")
plt.bar(xpos+0.2,profit, width=0.4,label="Profit")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()

# width to control the bars
```

## Barh for Horizontal Bars

It has no width argument

```python
# Horizontal Bars using barh
plt.barh(xpos,revenue, label="Revenue")

plt.yticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technolog Stocks')
plt.legend()
```
