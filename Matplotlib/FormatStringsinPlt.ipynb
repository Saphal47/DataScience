## Format Strings in Plot Function

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x= [1,2,3,4,5,6,7] #week days
y= [50,51,52,48,47,49,46]
```

```python
plt.plot(x,y,'g+') # + is marker and g for green
```

```python
plt.plot(x,y,'g+--')
```

```python
plt.plot(x,y,'g+-.') #dashdot line
```

```python
plt.plot(x,y,'--r*') #r for red
```

```python
plt.plot(x,y,'rD') #D for diamond
```

```python
#other way
plt.plot(x,y,color='red',marker='D',linestyle='')
```

```python
plt.plot(x,y,color='red',marker='+',linestyle='',markersize=16)
```

```python
#alpha for transperency
plt.plot(x,y,color='red',alpha=0.2)
```
