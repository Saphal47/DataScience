## Axis labels,Legend and Grid

```python
import matplotlib.pyplot as plt #pyplot is VVIP
%matplotlib inline
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[32,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.plot(days,max_t)
```

```python
plt.title("Weather")
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
#plot 3 in one
```

Applying Legend on above plot

```python
plt.title("Weather")
plt.xlabel("Days")
plt.ylabel("Temperature")

plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

plt.legend(loc="best", shadow=True , fontsize="small") 
#loc to change location of legend
#Shadow for nice UI of legend
```
