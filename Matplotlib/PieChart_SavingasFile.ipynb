## Pie Chart

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]

plt.pie(exp_vals,labels=exp_labels)
```

Draw pie chart as perfect circle

```python
plt.pie(exp_vals,labels=exp_labels)
plt.axis("equal")
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels,radius=1.5)
plt.show()
```

Show percentages for every pie.

Specify radius to increase chart size

```python
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.2f%%')
plt.show()
```

### Explode for the pie to come out of the chart

To take out pies.It is array like 

```python
plt.pie(exp_vals,labels=exp_labels,radius=1.5,explode=[0,0,0,0,0],autopct='%0.2f%%')
plt.show()
```

```python
# Five values in explode array as there are 5 pies
plt.pie(exp_vals,labels=exp_labels,radius=1.5,explode=[0,0.1,0,0,0],autopct='%0.2f%%')
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels,radius=1.5,explode=[0,0.5,0,0,0],autopct='%0.2f%%')
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels,radius=1.5,explode=[0,0.1,0.1,0,0],autopct='%0.2f%%')
plt.show()
```

## Change the start angle of pieChart

```python
# also can use numpy array for explode by importing csv
plt.pie(exp_vals,labels=exp_labels,radius=1.5,explode=[0,0,0,0,0],autopct='%0.2f%%', startangle=45)
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels,radius=1.5,explode=[0,0,0,0,0],autopct='%0.2f%%',startangle=90)
plt.show()
```

# Saving Chart to a File

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5,explode=[0,0.1,0.1,0,0],autopct='%0.2f%%',startangle=90)
plt.show()

plt.savefig("piechart.jpg",bbox_inches="tight",pad_inches=2,transparent=True)
```

```python

plt.pie(exp_vals,labels=exp_labels,radius=2, autopct='%0.1f%%', explode=[0,0.1,0.1,0,0])
plt.savefig("piechart.pdf", bbox_inches="tight",pad_inches=2)
```
